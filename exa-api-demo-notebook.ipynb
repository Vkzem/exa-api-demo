```python
# Import OS module to provide Exa API key as an environment variable
import os

# Set the Exa API key - find yours in your Exa dashboard overview
exa_api_key = os.getenv('EXA_API_KEY')
```

```python
# If required, install the exa_py module
# %pip install exa_py
```

```python
# Import Exa module
from exa_py import Exa

# Create an Exa instance
exa = Exa(exa_api_key)

# Run the Exa query with the search_and_contents method - this example searches for a specific kind of case and fetches some of the contents in the return results
response = exa.search_and_contents(
  "Most recent common law cases in Australia, covering Australian Consumer Law statute Section 18 (relating to, in trade or commerce, conduct that is misleading or deceptive or is likely to mislead or deceive)",
  num_results=10,
  use_autoprompt=True,
  highlights={"num_sentences": 3, },
)

# Print the response result
print(response)
```
